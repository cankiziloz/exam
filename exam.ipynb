```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('country_vaccination_stats.csv')
```

```python
df[:10]
```

```python
df[np.isnan(df['daily_vaccinations'])]
```

```python
min_vaccination_by_country = df.groupby('country')[['daily_vaccinations']].min()
```

```python
min_vaccination_by_country
```

```python
min_vaccination_by_country.fillna(0, inplace=True)
```

```python
min_vaccination_by_country
```

```python
df['daily_vaccinations'] = df.apply(lambda row: min_vaccination_by_country.loc[row['country']] if np.isnan(row['daily_vaccinations']) else row['daily_vaccinations'], axis=1)
```

```python
df[:10]
```

```python
df.groupby('country').median().sort_values('daily_vaccinations')[-3:]
```

```python
np.sum(df[df['date'] == '1/6/2021']['daily_vaccinations'])
```
